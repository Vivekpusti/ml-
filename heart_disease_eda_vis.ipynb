```python
#Observing the skewness and kurtowsis of the data
heart_data = data.drop(columns=['target'],axis=1)
target = data['target']
```

Categorical variables like cp, sex, slope, ca, thal and exang need to converted to one hot encoding as they are currently in label encoding even though they do not represent any kind of heirarchy.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
transformer = make_column_transformer((OneHotEncoder(),['cp','sex','exang','fbs','thal','ca','restecg','slope']),remainder='passthrough')
```

```python
heart_data = pd.DataFrame(transformer.fit_transform(heart_data),columns=transformer.get_feature_names())
```
